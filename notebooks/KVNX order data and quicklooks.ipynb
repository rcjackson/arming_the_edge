```python
my_file_list = sorted(glob(lcn + 'KVNX*'))
out_path = '/nfs/gce/projects/digr/kvnx_data/quicklooks'
```

```python
for file in my_file_list:
    radar = pyart.io.read(file)
    rad_time = datetime.datetime.strptime(radar.time["units"], "seconds since %Y-%m-%dT%H:%M:%SZ")
    fig, ax = plt.subplots(1, 1, figsize=(6,6))
    disp = pyart.graph.RadarMapDisplay(radar)
    disp.plot_ppi_map('reflectivity', sweep=0, ax=ax, cmap='pyart_HomeyerRainbow', vmin=0, vmax=60,
                      min_lat=34., max_lat=39., min_lon=-101, max_lon=-95)

    outpath2 = out_path + '/%04d/%02d/%02d/' % (rad_time.year, rad_time.month, rad_time.day)
    if not os.path.exists(outpath2):
        os.makedirs(outpath2)
    fig.savefig(outpath2 + '/KVNX%04d%02d%02d_%02d%02d%02d.png' % 
              (rad_time.year, rad_time.month, rad_time.day, rad_time.hour, rad_time.minute, rad_time.second),
               dpi=300)
    del radar, disp
    plt.close(fig)
```
